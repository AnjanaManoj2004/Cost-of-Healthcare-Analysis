```python
# Treatment Efficiency and Comparison – Draft Notebook
# Author: Anil K C

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Example simulated data
data = {
    'Treatment_Type': ['Surgery', 'Mental Health', 'General Medicine'],
    'Cost': [1200, 600, 400],
    'Outcome_Score': [80, 70, 60]
}

df = pd.DataFrame(data)

# Visualise
sns.set(style='whitegrid')
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='Cost', y='Outcome_Score', hue='Treatment_Type', s=100)
plt.title('Treatment Cost vs Outcome')
plt.xlabel('Cost (AUD)')
plt.ylabel('Outcome Score')
plt.show()
```
